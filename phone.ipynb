```python
'''columns which has missing values 
1.) num_cores -> 6 missing -> fill with mean value 
2.) processor_speed-> 42 missing value -> fill with mean value 
3.) battery capacity -> 11 missing value -> fill with mean value 
4.) num of front cameras ->4 missing values -> fill with mode value 
5.) os-> 14 missing values -> drop 
6.) primary_camera_front -> 5 missing values -> fill with mode value  '''


from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
imputer= SimpleImputer()

transformer= ColumnTransformer(transformers=[
    ('tnf1', SimpleImputer(),["processor_speed","battery_capacity","num_cores"]),
    ('tnf2',SimpleImputer(strategy='most_frequent'),['num_front_cameras',"primary_camera_front","os","processor_brand"]),

], remainder='passthrough',verbose_feature_names_out=False)


transformed= transformer.fit_transform(df.iloc[:,:])


```

```python
new_df=pd.DataFrame(
    transformed,
    columns=transformer.get_feature_names_out()
)
```

```python
brand_counts = new_df['brand_name'].value_counts()

# Brands having at least 10 phones
valid_brands = brand_counts[brand_counts >= 10].index

# Replace rare brands with "Others"
new_df['brand_name'] = new_df['brand_name'].apply(
    lambda x: x if x in valid_brands else 'Others'
)
new_df["brand_name"].value_counts()
```

```python
new_df["processor_brand"].value_counts()
```

```python
brand_counts = new_df['processor_brand'].value_counts()

# Brands having at least 10 phones
valid_brands = brand_counts[brand_counts >= 10].index

# Replace rare brands with "Others"
new_df['processor_brand'] = new_df['processor_brand'].apply(
    lambda x: x if x in valid_brands else 'Others'
)
new_df["processor_brand"].value_counts()
```

```python
new_df["has_5g"] = new_df["has_5g"].apply(
    lambda x: 1 if x == True else 0
)
new_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe_transformer= ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(drop="first",sparse_output=False,dtype='int'),["brand_name","os","processor_brand"])
],remainder="passthrough",verbose_feature_names_out=False)

ohe_transformed= ohe_transformer.fit_transform(new_df.iloc[:,:])
new_df= pd.DataFrame(ohe_transformed,columns=ohe_transformer.get_feature_names_out())

new_df.head()
```

```python
# processor_speed -> float
new_df["processor_speed"] = pd.to_numeric(
    new_df["processor_speed"],
    errors="coerce"
).astype(float)

# remaining columns -> int
for col in new_df.columns:
    if col != "processor_speed":
        new_df[col] = pd.to_numeric(
            new_df[col],
            errors="coerce"
        ).astype(int)
```

```python
new_df["price"] = np.log1p(new_df["price"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_df.drop(columns=["price"]),new_df["price"], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rfr= RandomForestRegressor(max_depth=5)
model_rfr.fit(X_train,y_train)
```

```python
y_pred_rfr= model_rfr.predict(X_test)
```

```python
r2_score(y_train,model_rfr.predict(X_train))
```

```python
r2_score(y_test,y_pred_rfr)
```
